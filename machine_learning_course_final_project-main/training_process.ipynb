# Training
 This Notebook uses processed data from eda.ipynb to train machine learning models

```python
import pandas as pd 
import os 

dataset_path = os.path.join(os.getcwd(), "dataset/processed_dataset.csv")
dataset_preprocessed = pd.read_csv(dataset_path)

try:
    dataset_preprocessed.drop(["Unnamed: 0"], axis=1, inplace=True)
    
except:
    pass 

dataset_preprocessed.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC 
from sklearn.ensemble import RandomForestClassifier 
from sklearn.model_selection import GridSearchCV
from sklearn import metrics 

from sklearn.metrics import (
    accuracy_score,
    f1_score,
    precision_score, 
    recall_score, 
    confusion_matrix, 
    ConfusionMatrixDisplay, 

    )
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split, cross_val_score

import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
dataset_train, dataset_test = train_test_split(dataset_preprocessed, 
                                               test_size=0.2, 
                                               shuffle=True, 
                                               random_state=15)
print(f"Shape of train dataset is: {dataset_train.shape}")
print(f"Shape of test dataset is: {dataset_test.shape}")
```

```python
x_train, y_train = dataset_train.drop(["Survived"], axis=1), dataset_train["Survived"]
x_test, y_test = dataset_test.drop(["Survived"], axis=1), dataset_test["Survived"]

print(f"shape of x_train is: {x_train.shape}")
print(f"shape of x_test is: {x_test.shape}")
print(f"shape of y_train is: {y_train.shape}")
print(f"shape of y_test is: {y_test.shape}")
```

## Logistic Regression

```python
lr = LogisticRegression(max_iter=1000, random_state=42)
cv = KFold(n_splits=5, shuffle=True, random_state=42)
sfs = SequentialFeatureSelector(
    lr,
    n_features_to_select=3,
    direction="forward",     
    scoring="accuracy",
    cv=cv,
    n_jobs=-1
)

sfs.fit(x_train, y_train)

selected_features = sfs.get_support()
print(selected_features)

x_train_selected_lr = sfs.transform(x_train)

scores = cross_val_score(lr, x_train_selected_lr, y_train, cv=cv, scoring="accuracy")

print("Fold accuracies:", scores)
print("Mean accuracy:", scores.mean())

lr.fit(x_train_selected_lr, y_train)

feature_names = np.array(x_train.columns)
selected_feature_names = feature_names[sfs.get_support()]

print("Selected features:", selected_feature_names)
```

```python
x_test_selected_lr = sfs.transform(x_test)
preds = lr.predict(x_test_selected_lr)

acc_lr = accuracy_score(y_test, preds)
f1_lr = f1_score(y_test, preds)
precision_lr = precision_score(y_test, preds)
recall_lr = recall_score(y_test, preds)

print(f"accuracy is: {acc_lr}")
print(f"f1 is: {f1_lr}")
print(f"precision is: {precision_lr}")
print(f"recall is: {recall_lr}")
```

```python
cfd = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, preds), 
                       display_labels=lr.classes_)
cfd.plot()
plt.grid(False)
plt.show()
```

## KNN

```python
knn = KNeighborsClassifier(n_neighbors=3)
cv = KFold(n_splits=5, shuffle=True, random_state=42)
sfs = SequentialFeatureSelector(
    knn,
    n_features_to_select=3,
    direction="forward",     
    scoring="accuracy",
    cv=cv,
    n_jobs=-1
)

sfs.fit(x_train, y_train)

selected_features = sfs.get_support()
print(selected_features)

x_train_selected_knn = sfs.transform(x_train)

scores = cross_val_score(knn, x_train_selected_knn, y_train, cv=cv, scoring="accuracy")

print("Fold accuracies:", scores)
print("Mean accuracy:", scores.mean())

knn.fit(x_train_selected_knn, y_train)
```

```python
x_test_selected_knn = sfs.transform(x_test)
preds = knn.predict(x_test_selected_knn)

acc_knn = accuracy_score(y_test, preds)
f1_knn = f1_score(y_test, preds)
precision_knn = precision_score(y_test, preds)
recall_knn = recall_score(y_test, preds)

print(f"accuracy is: {acc_knn}")
print(f"f1 is: {f1_knn}")
print(f"precision is: {precision_knn}")
print(f"recall is: {recall_knn}")
```

```python
cfd = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, preds), 
                       display_labels=knn.classes_)
cfd.plot()
plt.grid(False)
plt.show()
```

## SVM

```python
svm = SVC(random_state=42)
cv = KFold(n_splits=5, shuffle=True, random_state=42)
sfs = SequentialFeatureSelector(
    svm,
    n_features_to_select=3,
    direction="forward",     
    scoring="accuracy",
    cv=cv,
    n_jobs=-1
)

sfs.fit(x_train, y_train)

selected_features = sfs.get_support()
print(selected_features)

x_train_selected_svm = sfs.transform(x_train)

scores = cross_val_score(svm, x_train_selected_svm, y_train, cv=cv, scoring="accuracy")

print("Fold accuracies:", scores)
print("Mean accuracy:", scores.mean())

svm.fit(x_train_selected_svm, y_train)
```

```python
x_test_selected_svm = sfs.transform(x_test)
preds = svm.predict(x_test_selected_svm)

acc_svm = accuracy_score(y_test, preds)
f1_svm = f1_score(y_test, preds)
precision_svm = precision_score(y_test, preds)
recall_svm = recall_score(y_test, preds)

print(f"accuracy is: {acc_svm}")
print(f"f1 is: {f1_svm}")
print(f"precision is: {precision_svm}")
print(f"recall is: {recall_svm}")
```

```python
cfd = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, preds), 
                       display_labels=svm.classes_)
cfd.plot()
plt.grid(False)
plt.show()
```

## Random Forest

```python
rf = RandomForestClassifier(random_state=42)
cv = KFold(n_splits=5, shuffle=True, random_state=42)
sfs = SequentialFeatureSelector(
    rf,
    n_features_to_select=3,
    direction="forward",     
    scoring="accuracy",
    cv=cv,
    n_jobs=-1
)

sfs.fit(x_train, y_train)

selected_features = sfs.get_support()
print(selected_features)

x_train_selected_rf = sfs.transform(x_train)

scores = cross_val_score(rf, x_train_selected_rf, y_train, cv=cv, scoring="accuracy")

print("Fold accuracies:", scores)
print("Mean accuracy:", scores.mean())

rf.fit(x_train_selected_rf, y_train)
```

```python
x_test_selected_rf = sfs.transform(x_test)
preds = rf.predict(x_test_selected_rf)

acc_rf = accuracy_score(y_test, preds)
f1_rf = f1_score(y_test, preds)
precision_rf = precision_score(y_test, preds)
recall_rf = recall_score(y_test, preds)

print(f"accuracy is: {acc_rf}")
print(f"f1 is: {f1_rf}")
print(f"precision is: {precision_rf}")
print(f"recall is: {recall_rf}")
```

```python
cfd = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix(y_test, preds), 
                       display_labels=rf.classes_)
cfd.plot()
plt.grid(False)
plt.show()
```

## Choosing Model

```python
precision_lr, recall_lr, thresholds = metrics.precision_recall_curve(y_test, lr.predict(x_test_selected_lr))
precision_knn, recall_knn, thresholds = metrics.precision_recall_curve(y_test, knn.predict(x_test_selected_knn))
precision_svm, recall_svm, thresholds = metrics.precision_recall_curve(y_test, svm.predict(x_test_selected_svm))
precision_rf, recall_rf, thresholds = metrics.precision_recall_curve(y_test, rf.predict(x_test_selected_rf))
```

```python
plt.figure(figsize=(6, 5))
plt.plot(precision_lr, recall_lr, label="logistic regression")
plt.plot(precision_knn, recall_knn, label="KNN")
plt.plot(precision_svm, recall_svm, label="SVM")
plt.plot(precision_rf, recall_rf, label="random forest")
plt.xlabel("Precision")
plt.ylabel("Recall")
plt.title("Precision Recall Curve")
plt.legend()
plt.grid(True)
plt.show()
```

```python
fpr_lr, tpr_lr, _ = metrics.roc_curve(y_test, lr.predict(x_test_selected_lr))
fpr_knn, tpr_knn, _ = metrics.roc_curve(y_test, knn.predict(x_test_selected_knn))
fpr_svm, tpr_svm, _ = metrics.roc_curve(y_test, svm.predict(x_test_selected_svm))
fpr_rf, tpr_rf, _ = metrics.roc_curve(y_test, rf.predict(x_test_selected_rf))

auc_lr = metrics.auc(fpr_lr, tpr_lr)
auc_knn = metrics.auc(fpr_knn, tpr_knn)
auc_svm = metrics.auc(fpr_svm, tpr_svm)
auc_rf = metrics.auc(fpr_rf, tpr_rf)

plt.figure(figsize=(6, 5))

plt.plot(fpr_lr, tpr_lr, label=f"LogReg (AUC={auc_lr:.2f})")
plt.plot(fpr_knn, tpr_knn, label=f"KNN (AUC={auc_knn:.2f})")
plt.plot(fpr_svm, tpr_svm, label=f"SVM (AUC={auc_svm:.2f})")
plt.plot(fpr_rf, tpr_rf, label=f"RF (AUC={auc_rf:.2f})")
plt.plot([0, 1], [0, 1], linestyle="--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.show()
```

Based on Precision Recall curve and ROC curve best model is random forest

## Hyper parameter tuning and saving model

```python
param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [None, 10, 20],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2]
}

cv = KFold(n_splits=5, shuffle=True, random_state=42)

grid = GridSearchCV(
    estimator=rf,
    param_grid=param_grid,
    cv=cv,
    scoring="accuracy",
    n_jobs=-1,
    verbose=1
)

grid.fit(x_train[["Pclass", "Sex", "SibSp"]].to_numpy(), y_train)
print("Best params:", grid.best_params_)
print("Best CV accuracy:", grid.best_score_)
best_rf = grid.best_estimator_

y_pred = best_rf.predict(x_test_selected_rf)
print("Test accuracy:", accuracy_score(y_test, y_pred))
```

```python
import pickle

with open("model/random_forest_model.pkl", "wb") as f:
    pickle.dump(best_rf, f)
```
